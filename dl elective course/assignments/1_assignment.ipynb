### Tensor Operations

```python
# tensor operations (creation, reshaping, broadcasting)

######################
# tensor creation
######################

# numpy tensor initialization
a_numpy = np.array([[1, 2, 3], [4, 5, 6]])
b_numpy =  np.array([[7, 8, 9], [10, 11, 12]])

# pytorch tensor initialization
a_torch = torch.tensor([[1, 2, 3], [4, 5, 6]])
b_torch = torch.tensor([[7, 8, 9], [10, 11, 12]])

print(f"\n numpy tensor: {a_numpy} \n")
print(f"\n torch tensor: {a_torch} \n")

######################
# Manual
######################

# mannual creation as a list
manual_tensor = [
    [1, 2, 3],
    [4, 5, 6]
]

print(f"\n manual tensor: {manual_tensor} \n")
```

```python
######################
# Tensor Reshaping
######################

a_numpy_tensor = np.array([1,2,3,4,5,6])

# Numpy tensor reshaped
reshaped_a_numpy = a_numpy_tensor.reshape(2,3)

a_torch_tensor = torch.tensor([1,2,3,4,5,6])

# Pytorch tensor reshaped
reshaped_a_torch = a_torch_tensor.reshape(2,3)

print(f"\n numpy_tensor a: {a_numpy_tensor} \n")
print(f"\n Reshaped Numpy Tensor: {reshaped_a_numpy} \n")
print(f"\n pytorch_tensor a: {a_torch_tensor} \n")
print(f"\n Reshaped Torch Tensor: {reshaped_a_torch} \n")

######################
# Manual
######################

# manual tensor
manual_tensor = [
    [1, 2, 3],
    [4, 5, 6]
]

# tranfer each manual tensor elements into a single list
flat_list = []
for row in manual_tensor:
    for elements in row:
        flat_list.append(elements)

# Reshape into 3x2 tensor
reshaped_manual = []
rows, cols = 3, 2
for i in range(rows): # range generate 0 to 2
    start_index = i*cols 
    end_index = start_index +cols
    row = flat_list[start_index:end_index]
    reshaped_manual.append(row)

print(f"\n manual_tensor a: {manual_tensor} \n")
print(f"\n Reshaped Manual Tensor: {reshaped_manual} \n")
```

```python
######################
# Tensor Broadcasting
######################
'''  
Broadcasting means operation between two tensor of different dimension. 
Simply, it means copying the smallest tensor to match the dimension of the largest tensor

Constraint: column of first tensor much be same of another or one should be 1 to be stretched

Here, first, I will try 2 different tensor 
initialized from pytorch and numpy 
with different dimension.
'''

a_numpy_tensor = np.array([[1,2,3],[4,5,6]])
b_pytorch_tensor = torch.tensor([10,20,30])
b_numpy_tensor = np.array([10,20,30])

# Not possible
# broadcasted_result = a_numpy_tensor + b_pytorch_tensor

broadcasted_result = a_numpy_tensor + b_numpy_tensor

print(f"\n Numpy a Tensor: {a_numpy_tensor} \n")
print(f"\n Numpy b Tensor: {b_numpy_tensor} \n")
print(f"\n broadcasted_result: {broadcasted_result} \n")

######################
# Manual
######################

# manual tensor
a_manual_tensor = [
    [1, 2, 3],
    [4, 5, 6]
]
b_manual_tensor = [10, 20, 30]
broadcasted_manual_result = []

# This is hard coded. This logic is not generic for other dimension tension.
for row in a_manual_tensor:
    new_row = []
    for i in range(len(row)):
        new_row.append(row[i] + b_manual_tensor[i]) # add element of smallest tensor to largest one
    broadcasted_manual_result.append(new_row)

print(f"\n Manual Tensor a: {a_manual_tensor} \n")
print(f"\n Manual Tensor b: {b_manual_tensor} \n")
print(f"\n broadcasted_result: {broadcasted_manual_result} \n")
```

## Part II

### Matrix Operation

```python
######################
# Matrix Operation (Multiplication)
######################
'''  
Matrix Multiplication can be done using @ opeartor which is Pythonic.
Also using matmul(a,b), works with higher dimension 
'''

# Numpy matrix definition
a_numpy = np.array([[1, 2, 3], 
                    [4, 5, 6]])
b_numpy =  np.array([[7, 8, 9], 
                     [10, 11, 12]])
numpy_operation = a_numpy @ b_numpy.T # Transposing to make valid size for multiplication

# Pytorch matrix definition
a_torch = torch.tensor([[1, 2, 3],
                        [4, 5, 6]])
b_torch = torch.tensor([[7, 8, 9], 
                        [10, 11, 12]])
pytorch_operation = a_torch @ b_torch.T


print(f"\n Numpy a Tensor: {a_numpy} \n")
print(f"\n Pytorch b Tensor: {b_torch} \n")
print(f"\n Numpy Matrix Muliplication Result: {numpy_operation} \n")
print(f"\n Pytorch Matrix Muliplication Result: {pytorch_operation} \n")
```

```python
######################
# Eigen Value
######################
'''
Pytorch return eigen value and vector in complex form to reduce error.
If is for safety reason and note to know that complex number are superset of real numbers
'''

# Numpy matrix definition
a_numpy = np.array([[1, 2, 3], 
                    [4, 5, 6],
                    [4, 5, 6]])
# Compute eigenvalues and eigenvectors
eigvals_numpy, eigvecs_numpy = np.linalg.eig(a_numpy)


# Pytorch matrix definition
a_torch = torch.tensor([[1, 2, 3],
                        [4, 5, 6],
                        [4, 5, 6]], dtype=torch.float32)
eigvals_torch, eigvecs_torch = torch.linalg.eig(a_torch)


print(f"\n Numpy Tensor a: {a_numpy} \n")
print(f"\n Pytorch Tensor a: {a_torch} \n")
print("\nNumPy Eigenvalues:\n", eigvals_numpy)
print("\nNumPy Eigenvectors:\n", eigvecs_numpy)
print("\nTorch Eigenvalues:\n", eigvals_torch)
print("\nTorch Eigenvectors:\n", eigvecs_torch)
```

### Numerical Instability and Precision Analysis

Computers can't represent all the numbers exactly. There is error especially in large sums,nearly equal number and matrix operation with very small or very large number due to round off.

```python
##############################################
# Numerical Precision Analysis
##############################################
''' 
This example shows how large number can loss the precesion when
data type is not set properly. For an example: float32 won't able to 
compute for precise solution for larger value. Below is the simple
example code.
'''

# Numpy scalar definition
a_numpy = np.array(1e10 + 1, dtype=np.float32)
b_numpy = np.array(1e10, dtype=np.float32)

difference_numpy = a_numpy - b_numpy

# Pytorch scalar definition
a_torch = torch.tensor(1e10 + 1, dtype=torch.float64)
b_torch = torch.tensor(1e10, dtype=torch.float64)

difference_torch = a_torch - b_torch

print(f"\n Numpy a: {a_numpy} \n Numpy b: {b_numpy} \n")
print(f"\nNumpy difference: {difference_numpy}\n")
print(f"\nDue to float32 data type the result is 0 which is not precise, the correct precise answer is 1\n")
print(f"\n Torch a: {a_torch} \n Torch b: {b_torch} \n")
print(f"\n Pytorch difference: {difference_torch} \n")
```

```python
##############################################
# Numerical Instability Analysis
##############################################
''' 
This example shows how large number can raise the instability 
in the calculation. Below is the simple example code of matrix
multiplication in which large value caused the instability.
'''

# Numpy matrix definition
a_numpy = np.array([[1e10,1],[1,1.0000001]]) # by default float64 datatype
b_numpy = np.array([[1,1],[1,1]])

result_numpy = np.linalg.inv(a_numpy) @ b_numpy

# Pytorch matrix definition
a_torch = torch.tensor([[1e10, 1], [1, 1.0000001]], dtype=torch.float32)
b_torch = torch.tensor([[1, 1], [1, 1]], dtype=torch.float32)

result_torch = torch.linalg.inv(a_torch) @ b_torch

print(f"\n Numpy a: {a_numpy} \n Numpy b: {b_numpy} \n")
print(f"\nNumpy Operation Result: {result_numpy}\n")
print(f"\nInstability in calculation due to data type difference as there is very small number difference in two output\n")
print(f"\n Torch a: {a_torch} \n Torch b: {b_torch} \n")
print(f"\n Pytorch Operation Result: {result_torch} \n")
```
